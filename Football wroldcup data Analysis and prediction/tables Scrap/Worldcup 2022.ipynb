```python
#zip conacdes two variables of a form loop.

all_tables = pd.read_html('https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')
dict_table = {}
for letter, i in zip (alpha,range (12,68,7)):
    df = all_tables[i]
    df.rename(columns={df.columns[1]: 'Team' },inplace=True)
    df.pop('Qualification')
    dict_table[f'Group {letter}'] = df
    
```

```python
dict_table['Group A']
```

```python
#putting it into another text file , the dictionary 
with open('table','wb') as output:
    pickle.dump(dict_table, output)
    
```
